```python
from nltk.corpus import stopwords 
stop_words = set(stopwords.words('english'))
```

```python
#Removing stop words from the data.

for i in range(len(train_data['text'])):
    temp = train_data['text'][i]
    temp = temp.split()
    sentence = ''
    for j in temp: 
        if j not in stop_words:
            #print('in')
            sentence = sentence + ' ' + j
    train_data['text'][i] = sentence
```

```python
#Removing '%20' that is supposed to be space between keywords.
#for values nan, use the first word of tweet as keyword.
for i in range(len(train_data['keyword'])):

    if type(train_data['keyword'][i]) == str:
        if '%20' in train_data['keyword'][i]:
            temp = train_data['keyword'][i].split('%20')
            temp = temp[0]+' '+temp[1]
            train_data['keyword'][i] = temp
    else:
        temp = train_data['text'][i].split()

        if len(temp) == 1:
            temp.append('<OOV>')
        train_data['keyword'][i] = temp[0]+' '+temp[1]
 
```

```python
train_X_text = train_data['text']
train_X_keyword = train_data['keyword']
train_Y = train_data['target']
```

```python
train_Y.nunique()
```

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words = 22500, oov_token = '<OOV>')
tokenizer.fit_on_texts(train_X_text)
tokenizer.fit_on_texts(train_X_keyword)
word_index = tokenizer.word_index
```

```python
word_index.items()
```

```python
#Load and prepare GloVe embeddings.
path_to_glove_file = "C:\\Users\Saransh agrawal\Anaconda_installation\Anaconda\Lib\site-packages\\tensorflow\python\keras\datasets\glove.6B.100d.txt"

embeddings_index = {}
with open(path_to_glove_file, encoding="utf8") as f:
    for line in f:
        word, coefs = line.split(maxsplit=1)
        coefs = np.fromstring(coefs, "f", sep=" ")
        embeddings_index[word] = coefs

print("Found %s word vectors." % len(embeddings_index))
```

```python
num_tokens = len(train_Y) + 2
embedding_dim = 100
hits = 0
misses = 0
embeddings_matrix = np.zeros((len(word_index)+1, embedding_dim));
# Prepare embedding matrix
embedding_matrix = np.zeros((num_tokens, embedding_dim))
for word, i in word_index.items():
    embedding_vector = embeddings_index.get(word);
    if embedding_vector is not None:
        embeddings_matrix[i] = embedding_vector;
```

```python
sequences_text = tokenizer.texts_to_sequences(train_X_text)
padded_text = tf.keras.preprocessing.sequence.pad_sequences(sequences_text, maxlen=20, padding='post', truncating='post')

split = int(0.2*len(train_X_text))
training_size = len(train_X_text)
test_sequences = padded_text[0:split]
training_sequences = padded_text[split:training_size]
test_labels = train_Y[0:split]
training_labels = train_Y[split:training_size]

training_padded = np.array(training_sequences)
training_labels = np.array(training_labels)
testing_padded = np.array(test_sequences)
testing_labels = np.array(test_labels)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(len(word_index)+1, 100, input_length=20, weights=[embeddings_matrix], trainable=False),

    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Conv1D(128, 5, activation='relu'),
    tf.keras.layers.Dropout(0.4),

    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128, return_sequences = True)),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, return_sequences = True)),
    tf.keras.layers.Dropout(0.3),

     tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64)),
     tf.keras.layers.Dropout(0.2),

    tf.keras.layers.Dense(64, activation = 'relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
model.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy'])
model.summary()


```

```python
cb = tf.keras.callbacks.EarlyStopping(patience = 15, restore_best_weights=True)
history = model.fit(training_padded, training_labels, epochs=50, validation_data=(testing_padded, testing_labels), verbose=2,batch_size = 128, workers = 4, callbacks = [cb])
history2 =model.fit(testing_padded, testing_labels, epochs=17, validation_data=(training_padded, training_labels), verbose=2, batch_size = 64, workers = 4)
```

```python

sequences_key = tokenizer.texts_to_sequences(train_X_keyword)
padded_key = tf.keras.preprocessing.sequence.pad_sequences(sequences_key, maxlen=2, padding='post', truncating='post')

split = int(0.2*len(train_X_keyword))
training_size = len(train_X_keyword)
test_sequences = padded_key[0:split]
training_sequences = padded_key[split:training_size]
test_labels = train_Y[0:split]
training_labels = train_Y[split:training_size]

training_padded_key = np.array(training_sequences)
training_labels = np.array(training_labels)
testing_padded_key = np.array(test_sequences)
testing_labels = np.array(test_labels)
```

```python
#Model trained just on keywords.
model2 = tf.keras.Sequential([
    tf.keras.layers.Embedding(len(word_index)+1, 100, input_length=2, weights=[embeddings_matrix], trainable=False),
    tf.keras.layers.BatchNormalization(),

    tf.keras.layers.GlobalAveragePooling1D(),

    tf.keras.layers.Dense(64, activation = 'relu'),

    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(32, activation = 'relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])

model2.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model2.summary()
```

```python
cb = tf.keras.callbacks.EarlyStopping(patience = 20, restore_best_weights=True)
history = model2.fit(training_padded_key, training_labels, epochs=200, validation_data=(testing_padded_key, testing_labels), verbose=2,batch_size = 128, workers = 4, callbacks = [cb])
#history2 =model2.fit(testing_padded_key, testing_labels, epochs=21, validation_data=(training_padded_key, training_labels), verbose=2, batch_size = 64, workers = 4, callbacks = [cb])
```

```python
input1 = tf.keras.layers.Input(shape = [20])
input2 = tf.keras.layers.Input(shape = [2])
hidden1 = model(input1)
hidden2 = model2(input2)
concat = tf.keras.layers.Concatenate()([hidden1, hidden2])
output = tf.keras.layers.Dense(1, activation = 'sigmoid')(concat)
model_final = tf.keras.Model(inputs = [input1, input2], outputs = [output])

model_final.summary()
```

```python
model_final.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['acc'])
```
